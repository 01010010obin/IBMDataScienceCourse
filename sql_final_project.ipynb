### Problem 3

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%%sql SELECT *
FROM CHICAGO_CRIME_DATA 
WHERE LOWER(DESCRIPTION) LIKE '%child%' and lower(PRIMARY_TYPE) like "%kidnapping%"
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
%%sql SELECT *
FROM CHICAGO_CRIME_DATA 
WHERE LOWER(LOCATION_DESCRIPTION) LIKE '%school%' 
```

```python
%%sql SELECT distinct PRIMARY_TYPE
FROM CHICAGO_CRIME_DATA 
WHERE LOWER(LOCATION_DESCRIPTION) LIKE '%school%' 
```

```python
%%sql SELECT distinct PRIMARY_TYPE,DESCRIPTION
FROM CHICAGO_CRIME_DATA 
WHERE LOWER(LOCATION_DESCRIPTION) LIKE '%school%'  group by PRIMARY_TYPE,DESCRIPTION
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
%%sql SELECT distinct *
FROM CHICAGO_CRIME_DATA limit 5
```

```python
%%sql SELECT  COMMUNITY_AREA_NUMBER, count(COMMUNITY_AREA_NUMBER) as can
FROM CHICAGO_CRIME_DATA  group by COMMUNITY_AREA_NUMBER having COMMUNITY_AREA_NUMBER<>'None' order by can  desc limit 10
```

```python
%%sql
SELECT COMMUNITY_AREA_NUMBER, 
       COUNT(COMMUNITY_AREA_NUMBER) AS can, 
       (COUNT(COMMUNITY_AREA_NUMBER) * 1.0 / SUM(COUNT(COMMUNITY_AREA_NUMBER)) OVER ()) * 100 AS percentcan
FROM CHICAGO_CRIME_DATA
GROUP BY COMMUNITY_AREA_NUMBER
HAVING COMMUNITY_AREA_NUMBER <> 'None'
ORDER BY can DESC
LIMIT 10;
```

```python
%%sql
SELECT count(*)
FROM CHICAGO_CRIME_DATA
```

```python
%%sql
SELECT count(*)
FROM CHICAGO_CRIME_DATA where COMMUNITY_AREA_NUMBER!='None'
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql select cd.COMMUNITY_AREA_NAME,cc.COMMUNITY_AREA_NUMBER,count(cd.COMMUNITY_AREA_NAME) as cnt
    from CHICAGO_CRIME_DATA cc,CENSUS_DATA cd 
    where cd.COMMUNITY_AREA_NUMBER=cc.COMMUNITY_AREA_NUMBER group by cd.COMMUNITY_AREA_NAME,cc.COMMUNITY_AREA_NUMBER order by cnt desc limit 15
```

```python
%%sql select cd.COMMUNITY_AREA_NUMBER,cc.COMMUNITY_AREA_NUMBER,cd.COMMUNITY_AREA_NAME 
    from CHICAGO_CRIME_DATA cc,CENSUS_DATA cd 
    where cd.COMMUNITY_AREA_NUMBER=cc.COMMUNITY_AREA_NUMBER and cd.COMMUNITY_AREA_NAME='Brighton Park'
```

```python
%%sql SELECT COMMUNITY_AREA_NAME
FROM CENSUS_DATA
WHERE COMMUNITY_AREA_NUMBER in (SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC limit 10);
```

```python
%%sql SELECT cd.COMMUNITY_AREA_NUMBER, cd.COMMUNITY_AREA_NAME, COUNT(cc.COMMUNITY_AREA_NUMBER)
FROM CHICAGO_CRIME_DATA cc
JOIN CENSUS_DATA cd ON cc.COMMUNITY_AREA_NUMBER = cd.COMMUNITY_AREA_NUMBER
GROUP BY cd.COMMUNITY_AREA_NAME
ORDER BY COUNT(cc.COMMUNITY_AREA_NUMBER) DESC;
```
